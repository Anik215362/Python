# Class Work

Write a program for the following statement:

age = take user input

if

  age greater than 0 and less than 13, Print "Child"

  age greater than or equal 13 and less than 18, print"Teen"

  age greater than or equal 18 and less than or equal 35, print "Young"

  age greater than or equal 36 and less then 100, print "Old"

  OR print "Error" 

```python
age = int(input("Enter Age : "))

if age > 0 and age < 13:
  print("Child")
elif age >= 13 and age < 18:
  print("Teen")
elif age > 18 and age <= 35:
  print("Young")
elif age >= 36 and age <100:
  print("Old")
else:
  print("Error")
```

Write a program to check Odd or even number.

```python
n = int(input("Enter an interger: "))

if (n % 2) == 0:
  print("Even Number")
else:
  print("Odd Number")
```

Nested loop

```python
for i in range(11):
  for j in range(i):
    print('*', end='')
  print('')
```

Break Statement

```python
for number in range(10):
  if number==7:
    break
  print(number, end=" ")
```

Continue statement

```python
for number in range(10):
  if number==5:
    continue
  print(number, end=" ")
```
